```r
Ch.means = rowMeans(Ch)
head(Ch.means)
```

```r
H.means = rowMeans(H)
head(H.means)
```

```r
F.means = rowMeans(F)
head(F.means)
```

```r
plot(H.means ~ F.means)
abline(0, 1, col = "red")
```

```r
plot(H.means ~ Ch.means)
abline(0, 1, col = "red")
```

```r
HF_DIFF = F.means - H.means
head(HF_DIFF)
```

```r
HCh_DIFF = Ch.means - H.means
head(HCh_DIFF)
```

```r
hist(HF_DIFF, col = "gray")
```

```r
hist(HCh_DIFF, col = "gray")
```

```r
number_of_genes <- nrow(H)

pvalue1 = numeric(number_of_genes)

for(i in 1:number_of_genes) {
  x = H[i, ]
  y = Ch[i, ]

  if (var(x) == 0 || var(y) == 0) {
    pvalue1[i] = 1
  } else {
    t = t.test(x, y)
    pvalue1[i] = t$p.value
  }
}
head(pvalue1)
```

```r
length(pvalue1)
```

```r
number_of_genes <- nrow(H)

pvalue2 = numeric(number_of_genes)

for(i in 1:number_of_genes) {
  x2 = H[i, ]
  y2 = F[i, ]

  if (var(x2) == 0 || var(y2) == 0) {
    pvalue2[i] = 1
  } else {
    t = t.test(x2, y2)
    pvalue2[i] = t$p.value
  }
}
head(pvalue2)
```

```r
length(pvalue2)
```

```r
plot(-log10(pvalue1) ~ HCh_DIFF)
```

```r
plot(-log10(pvalue2) ~ HF_DIFF)
```

```r
HCh_DIFF_cutoff = 1
pvalue1_cutoff = 0.01

plot(-log10(pvalue1) ~ HCh_DIFF)

abline(v = HCh_DIFF_cutoff, col = "blue", lwd = 3)
abline(v = -HCh_DIFF_cutoff, col = "red", lwd = 3)
abline(h = -log10(pvalue1_cutoff), col = "green", lwd = 3)
```

```r
HF_DIFF_cutoff = 1
pvalue2_cutoff = 0.01

plot(-log10(pvalue2) ~ HF_DIFF)

abline(v = HF_DIFF_cutoff, col = "blue", lwd = 3)
abline(v = -HF_DIFF_cutoff, col = "red", lwd = 3)
abline(h = -log10(pvalue2_cutoff), col = "green", lwd = 3)
```

```r
filter_by_HF_DIFF = abs(HF_DIFF ) >=HF_DIFF_cutoff # Biological
head(filter_by_HF_DIFF)
```

```r
filter_by_HCh_DIFF = abs(HCh_DIFF ) >=HCh_DIFF_cutoff # Biological
head(filter_by_HCh_DIFF)
```

```r
sum(filter_by_HF_DIFF)
```

```r
sum(filter_by_HCh_DIFF)
```

```r
filter_by_pvalue1 = pvalue1 <= pvalue1_cutoff # Statistical
sum(filter_by_pvalue1)
```

```r
filter_by_pvalue2 = pvalue2 <= pvalue2_cutoff
sum(filter_by_pvalue2)
```

```r
filter_combinedCh = filter_by_HCh_DIFF & filter_by_pvalue1
sum(filter_combinedCh)
```

```r
filter_combinedF = filter_by_HF_DIFF & filter_by_pvalue2
sum(filter_combinedF)
```

```r
filteredCh = data2[filter_combinedCh, ]
dim(filteredCh)
```

```r
filteredF = data2[filter_combinedF, ]
dim(filteredF)
```

```r
dataPCA= read_tsv("https://github.com/ahmedmoustafa/gene-expression-datasets/raw/main/datasets/primates_diet/primates_diet.tsv")
```

```r
glimpse(dataPCA)
```

```r
set.seed(123456)

train = slice_sample(dataPCA %>% group_by(HumanCafeDiet2.3),
                     prop = 0.95,
                     replace = FALSE) %>%
  arrange(ChimpanzeeDiet2.3)

glimpse(train)
```

```r
test = dataPCA[-train$ChimpanzeeDiet2.3, ]
glimpse(test)
```

```r
PCA_result = train %>% ungroup() %>% select(-ChimpanzeeDiet2.3, -HumanCafeDiet2.3)
head(PCA_result)
```

```r

str(PCA_result)
PCA_result <- as.data.frame(lapply(PCA_result, function(x) {
  if (is.factor(x)) {
    as.numeric(as.character(x))
  } else {
    x
  }
}))

PCA_result <- PCA_result[, sapply(PCA_result, is.numeric)]

PCA_result <- na.omit(PCA_result)

pca <- prcomp(PCA_result, scale. = TRUE)
summary(pca)
```

```r
plot(pca)
```

```r
t1 = train %>% add_column (PC1 = pca$x[, 1], PC2 = pca$x[, 2])
glimpse(t1)
```
